## Selecting certain types of data

* What if we want all incidents of theft/larceny?
* We can return if a row matches a value or not
* We can then use this data frame to get rows that match it

```python
is_theft_df = crimes['Type'] == 'THEFT/LARCENY'
is_theft_df
```

```python
theft_df = crimes[is_theft_df]
theft_df
```

```python
# We can combine multiple conditions
is_100 = crimes['Score'] == 100
theft_df_100 = crimes[is_theft_df & is_100][:10]
theft_df_100
```

```python
# And we can just select columns out of this
theft_df_100[['BlockAddress', 'CMAgency', 'Date']]
```

```python
# Which type of crime has the most incidents of score 100?
high_score_incidents = crimes[crimes['Score'] == 100]
high_score_incidents['Type'].value_counts()
```

```python
# How about we compare this with the total number of incidents
percent_high_score = high_score_incidents['Type'].value_counts() / crimes['Type'].value_counts()
percent_high_score
```

```python
percent_high_score.plot(kind='bar')
```

```python
# Focusing on thefts

thefts = crimes[crimes['Type'] == 'THEFT/LARCENY'][['Date', 'BlockAddress', 'CMAgency', 'Score']]
thefts
```

```python
# Convert the dates to proper date times
thefts.info()
thefts['Date'] = thefts['Date'].astype('datetime64')
thefts
```

```python
# Get the weekdays
weekdays = thefts['Date'].dt.weekday
```

```python
thefts['Date'].dt.weekday
```

```python
thefts.loc[:, 'weekday'] = weekdays
```

```python
thefts
```

```python
# Group by weekday
counts = thefts.groupby('weekday').size()
counts
```

```python
# Add in an index with the names of the days
counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
counts
```

```python
counts.plot(kind='bar')
```

```python
# See which days have the highest average
score_average = thefts[['Score', 'weekday']].groupby('weekday').agg(['mean'])
score_average
```

```python
score_average.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
score_average
```

```python
100 - score_average
```

```python
(100 - score_average).plot()
```

```python
# String operations, back to crimes
crimes
```

```python
addresses = crimes['BlockAddress']
university_crimes = addresses.str.contains('UNIVERSITY')
dole_crimes = addresses.str.contains('DOLE')
university_area_crimes = crimes[university_crimes | dole_crimes][['Date', 'BlockAddress', 'Type']]
university_area_crimes
```

```python
university_area_crimes['Type'].value_counts()
```

```python
university_area_crimes = university_area_crimes.set_index('Date')
university_area_crimes.index = pd.to_datetime(university_area_crimes.index)
```

```python
university_area_crimes.resample('1D')['Type'].count().plot()
```
